```python
# Imports
import pandas as pd
```

```python
# helpers
def delete_accented_chars(text):
    text.replace('á', 'a').replace('é', 'e').replace('í', 'i').replace('ó', 'o').replace('ú', 'u')
    return text
```

```python
# Preprocessing class
class Preprocessing:

  def __init__(self):
    pass

  def preprocess(self, data: pd.DataFrame) -> pd.DataFrame:
    copy = data.copy()
    # Column working as true or false
    copy['working'] = copy['working'].astype('str').str.lower().apply(delete_accented_chars)
    copy['working'] = copy['working'].apply(lambda x: 1 if x == 'si' else 0)

    return copy
    
pp = Preprocessing()
```

```python
# import the dataset 
df = pd.read_csv('../data/dataset-v1.csv', delimiter=',', encoding='utf-8')
df
```

```python
# preprocess
df_preprocessed = pp.preprocess(df)
print(df_preprocessed.head())
```
